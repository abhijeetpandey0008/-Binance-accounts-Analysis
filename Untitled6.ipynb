```python
# Step 1: Data Exploration and Cleaning
import pandas as pd
import numpy as np
import ast
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import plotly.express as px
```

```python
# Set up visualization style
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (12, 6)

# Load the CSV file
df = pd.read_csv('/content/drive/MyDrive/TRADES_CopyTr_90D_ROI.csv')

# Inspect the first few rows
print("First few rows of the dataset:")
display(df.head())

# Check for missing values in key columns
missing_counts = {
    "Port_IDs": df['Port_IDs'].isnull().sum(),
    "Trade_History": df['Trade_History'].isnull().sum()
}
print("Missing values in key columns:", missing_counts)

```

```python
# Visualize missing values
plt.figure()
sns.barplot(x=list(missing_counts.keys()), y=list(missing_counts.values()))
plt.title("Missing Values in Key Columns")
plt.ylabel("Count")
plt.show()
```

```python
# Parse the Trade_History column from string to list of dictionaries.
def parse_trade_history(trade_history_str):
    try:
        trades = ast.literal_eval(trade_history_str)
        return trades if isinstance(trades, list) else []
    except Exception as e:
        print("Error parsing trade history:", e)
        return []
```

```python
# Apply the parsing function and create a new column 'Parsed_Trades'
df['Parsed_Trades'] = df['Trade_History'].apply(parse_trade_history)

# Create a new column with the count of trades per account
df['Trade_Count'] = df['Parsed_Trades'].apply(lambda x: len(x) if isinstance(x, list) else 0)

# Visualize the distribution of trade counts per account
plt.figure()
sns.histplot(df['Trade_Count'], bins=30, kde=True)
plt.title("Distribution of Trade Counts per Account")
plt.xlabel("Number of Trades")
plt.ylabel("Frequency")
plt.show()
```

```python
# (Optional) Visualize a sample account's trade timeline (cumulative PnL over time)
sample_account = df.iloc[0]
sample_trades = sample_account['Parsed_Trades']
if sample_trades:
    # Sort trades by time
    sample_trades_sorted = sorted(sample_trades, key=lambda x: x.get('time', 0))
    times = [datetime.fromtimestamp(trade.get('time', 0)/1000) for trade in sample_trades_sorted]
    profits = [float(trade.get('realizedProfit', 0)) for trade in sample_trades_sorted]
    cumulative = np.cumsum(profits)

    plt.figure()
    plt.plot(times, cumulative, marker='o')
    plt.title(f"Cumulative PnL Timeline for Account {sample_account['Port_IDs']}")
    plt.xlabel("Time")
    plt.ylabel("Cumulative PnL")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
## Step 2: Feature Engineering  Calculate Metrics for Each Account

# %% [code]
def calculate_metrics(trades):
    # Return default metrics if no trades exist
    if not trades or len(trades) == 0:
        return {
            'Total_Positions': 0,
            'Win_Positions': 0,
            'Win_Rate': np.nan,
            'PnL': 0,
            'Total_Invested': 0,
            'ROI': np.nan,
            'Sharpe': 0,
            'MDD': 0,
            'Cumulative_PnL_Series': []  # For optional visualization
        }

    total_positions = len(trades)
    win_positions = 0
    pnl = 0.0
    total_invested = 0.0
    returns = []  # For Sharpe ratio calculation

    # Sort trades by timestamp
    try:
        trades_sorted = sorted(trades, key=lambda x: x.get('time', 0))
    except Exception as e:
        trades_sorted = trades

    cumulative_pnl = []
    cum_sum = 0.0

    for trade in trades_sorted:
        realized_profit = trade.get('realizedProfit')
        quantity = trade.get('quantity')
        if realized_profit is None or quantity is None:
            continue

        try:
            realized_profit = float(realized_profit)
            quantity = float(quantity)
        except:
            continue

        pnl += realized_profit
        total_invested += quantity

        if realized_profit > 0:
            win_positions += 1

        if quantity != 0:
            trade_return = realized_profit / quantity
            returns.append(trade_return)

        cum_sum += realized_profit
        cumulative_pnl.append(cum_sum)

    win_rate = win_positions / total_positions if total_positions > 0 else np.nan
    ROI = (pnl / total_invested * 100) if total_invested != 0 else np.nan


    # Calculate Sharpe Ratio using trade returns
    if returns:
        returns_arr = np.array(returns)
        mean_return = np.mean(returns_arr)
        std_return = np.std(returns_arr)
        sharpe = mean_return / std_return * np.sqrt(len(returns_arr)) if std_return != 0 else 0
    else:
        sharpe = 0

    # Calculate Maximum Drawdown (MDD)
    if cumulative_pnl:
        cumulative_pnl_arr = np.array(cumulative_pnl)
        running_max = np.maximum.accumulate(cumulative_pnl_arr)
        drawdowns = running_max - cumulative_pnl_arr
        mdd = np.max(drawdowns)
    else:
        mdd = 0

    return {
        'Total_Positions': total_positions,
        'Win_Positions': win_positions,
        'Win_Rate': win_rate,
        'PnL': pnl,
        'Total_Invested': total_invested,
        'ROI': ROI,
        'Sharpe': sharpe,
        'MDD': mdd,
        'Cumulative_PnL_Series': cumulative_pnl
    }

# Calculate metrics for each account and store the results
metrics_list = []
for idx, row in df.iterrows():
    port_id = row['Port_IDs']
    trades = row['Parsed_Trades']
    metrics = calculate_metrics(trades)
    metrics['Port_IDs'] = port_id
    metrics_list.append(metrics)

metrics_df = pd.DataFrame(metrics_list)




```

```python
# Visualize distributions for key metrics
metric_cols = ['ROI', 'PnL', 'Sharpe', 'MDD', 'Win_Rate']
for col in metric_cols:
    plt.figure()
    sns.histplot(metrics_df[col].dropna(), bins=30, kde=True)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

```python
# Visualize the cumulative PnL series for one sample account
sample_metrics = metrics_df[metrics_df['Total_Positions'] > 0].iloc[0]
cumulative_series = sample_metrics['Cumulative_PnL_Series']
if cumulative_series:
    plt.figure()
    plt.plot(cumulative_series, marker='o')
    plt.title(f"Cumulative PnL Series for Account {sample_metrics['Port_IDs']}")
    plt.xlabel("Trade Number")
    plt.ylabel("Cumulative PnL")
    plt.show()
```

```python
## Step 3: Ranking Algorithm  Normalization and Composite Scoring

# %% [code]
def min_max_normalize(series):
    min_val = series.min()
    max_val = series.max()
    if max_val == min_val:
        return series.apply(lambda x: 0.5)
    return (series - min_val) / (max_val - min_val)

# Create normalized columns for the key metrics
metrics_df['norm_ROI'] = min_max_normalize(metrics_df['ROI'])
metrics_df['norm_PnL'] = min_max_normalize(metrics_df['PnL'])
metrics_df['norm_Sharpe'] = min_max_normalize(metrics_df['Sharpe'])
metrics_df['norm_MDD'] = min_max_normalize(metrics_df['MDD'])
metrics_df['norm_WinRate'] = min_max_normalize(metrics_df['Win_Rate'])

# Visualize normalized metrics with a boxplot for a quick overview
norm_cols = ['norm_ROI', 'norm_PnL', 'norm_Sharpe', 'norm_MDD', 'norm_WinRate']
plt.figure()
sns.boxplot(data=metrics_df[norm_cols])
plt.title("Boxplot of Normalized Metrics")
plt.show()

# Define the weights for each metric
weights = {
    'ROI': 0.3,
    'PnL': 0.2,
    'Sharpe': 0.2,
    'MDD': 0.1,   # Inverted below because lower is better
    'Win_Rate': 0.2
}

# Compute the composite score for each account
metrics_df['Score'] = (
    weights['ROI'] * metrics_df['norm_ROI'] +
    weights['PnL'] * metrics_df['norm_PnL'] +
    weights['Sharpe'] * metrics_df['norm_Sharpe'] +
    weights['MDD'] * (1 - metrics_df['norm_MDD']) +
    weights['Win_Rate'] * metrics_df['norm_WinRate']
)

# Sort the accounts by composite score (highest first)
metrics_df.sort_values('Score', ascending=False, inplace=True)
metrics_df.reset_index(drop=True, inplace=True)

# Visualize the distribution of the composite score
plt.figure()
sns.histplot(metrics_df['Score'], bins=30, kde=True)
plt.title("Distribution of Composite Score")
plt.xlabel("Composite Score")
plt.ylabel("Frequency")
plt.show()
```

```python
# Additional Visualizations: Correlation Heatmap & Scatter Plots
corr_matrix = metrics_df[['ROI', 'PnL', 'Sharpe', 'MDD', 'Win_Rate']].corr()
plt.figure()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of Key Metrics")
plt.show()

# Interactive scatter plot: ROI vs PnL using Plotly
fig = px.scatter(metrics_df, x="ROI", y="PnL", hover_data=["Port_IDs"],
                 title="Interactive Scatter Plot: ROI vs PnL")
fig.show()
```
